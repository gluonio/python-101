# Tipe Data

```python
tipe_data_integer = 26
print(f'Data dalam variabel tipe_data_integer adalah {tipe_data_integer}, tipe datanya adalah {type(tipe_data_integer)}')
```

```python
tipe_data_float = 2.6
print(f'Data dalam variabel tipe_data_float adalah {tipe_data_float}, tipe datanya adalah {type(tipe_data_float)}')
```

```python
tipe_data_complex = 1+2j
print(f'Data dalam variabel tipe_data_complex adalah {tipe_data_complex}, tipe datanya adalah {type(tipe_data_complex)}')
```

```python
tipe_data_string = 'Juni'
print(f'Data dalam variabel tipe_data_string adalah {tipe_data_string}, tipe datanya adalah {type(tipe_data_string)}')
```

```python
tipe_data_list = [26, 'Juni', 2026, 17.31]
print(f'Data dalam variabel tipe_data_list adalah {tipe_data_list}, tipe datanya adalah {type(tipe_data_list)}')
```

```python
tipe_data_tuple = (26, 'Juni', 2026, 17.32)
print(f'Data dalam variabel tipe_data_tuple adalah {tipe_data_tuple}, tipe datanya adalah {type(tipe_data_tuple)}')
```

```python
tipe_data_set = {26, 27, 26, 26, 26, 28, 29, 30}
print(f'Data dalam variabel tipe_data_set adalah {tipe_data_set}, tipe datanya adalah {type(tipe_data_set)}')
```

```python
tipe_data_dictionary = {
    'tanggal' : 26,
    'bulan' : 'Juni',
    'tahun' : 2026,
    'waktu' : 17.35
}
print(f'Data dalam variabel tipe_data_dictionary adalah {tipe_data_dictionary}, tipe datanya adalah {type(tipe_data_dictionary)}')
```
